# Removing outliers using Inter Quartile Range

```python
gaming = df[df['Industry'] == 'Gaming']
Q1 = gaming['Human-AI Collaboration Rate (%)'].quantile(0.25)
Q3 = gaming['Human-AI Collaboration Rate (%)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df.loc[df['Industry'] == 'Gaming', 'Human-AI Collaboration Rate (%)'] = \
    df.loc[df['Industry'] == 'Gaming', 'Human-AI Collaboration Rate (%)'].clip(lower, upper)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Industry', y='Human-AI Collaboration Rate (%)')
plt.xticks(rotation=90);
```

```python
education = df[df['Industry'] == 'Education']
Q1 = education['Job Loss Due to AI (%)'].quantile(0.25)
Q3 = education['Job Loss Due to AI (%)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR


df.loc[df['Industry']== 'Education', 'Job Loss Due to AI (%)'] =\
           df.loc[df['Industry']== 'Education', 'Job Loss Due to AI (%)'].clip(lower, upper)
```

```python
healthcare = df[df['Industry'] == 'Healthcare']
Q1 = healthcare['Job Loss Due to AI (%)'].quantile(0.25)
Q3 = healthcare['Job Loss Due to AI (%)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR


df.loc[df['Industry']== 'Healthcare', 'Job Loss Due to AI (%)'] =\
           df.loc[df['Industry']== 'Healthcare', 'Job Loss Due to AI (%)'].clip(lower, upper)
```

```python
finance_industry = df[df['Industry'] == 'Finance']
Q1 = finance_industry['Job Loss Due to AI (%)'].quantile(0.25)
Q3 = finance_industry['Job Loss Due to AI (%)'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR


df.loc[df['Industry']== 'Finance', 'Job Loss Due to AI (%)'] =\
           df.loc[df['Industry']== 'Finance', 'Job Loss Due to AI (%)'].clip(lower, upper)
```

# Feature Engineering

```python
df.drop('Regulation Status', axis=1, inplace=True)
```

```python
df
```

```python
df_new=df[df['Year'] != 2025]
df_new
```

```python
df1=pd.get_dummies(df_new, columns=['Industry','Country', 'Top AI Tools Used'], drop_first=True)
df1
```

```python
df1['Human-AI Collaboration Rate (%)'] = df1['Human-AI Collaboration Rate (%)'].round(2)
```

```python
df1=df1.reset_index(drop=True)
df1
```

```python
numeric_col = df1.select_dtypes(include=['int64','float64']).columns
numeric_col
```

```python
df_norm=df1[numeric_col]
df_norm
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_array= scaler.fit_transform(df_norm)
scaled_df = pd.DataFrame(scaled_array, columns=df_norm.columns, index=df_norm.index)
scaled_df
```

```python
encoded_df = df1.select_dtypes(include=['uint8']).columns
encoded_df
```

```python
df2=df1[encoded_df]
df2
```

```python
df_combined = pd.concat([scaled_df,df2], axis=1)
df_combined
```

# Check for Multicollinearity

```python
X=df_combined.drop(columns=['Market Share of AI Companies (%)'])
y=df_combined['Market Share of AI Companies (%)']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_df = pd.DataFrame()
vif_df['Features']= X.columns
vif_df['VIF']= [round(variance_inflation_factor(X.values,i),2) for i in range(X.shape[1])]
vif_df
```

## we can see the VIF values are below 5, which means the data is showing moderate multicollinearity, which is sufficient enough to proceed with Linear Regression.

```python
numeric_col = scaled_df.select_dtypes(include=np.number).columns
for col in numeric_col:
    plt.figure(figsize=(4,2))
    sns.boxplot(x=scaled_df[col])
    plt.title(f"outliers in {col}")
    plt.show()
```

# Linear Regression

**Considered "Market Share of AI companies(%)" as Target variable**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
residuals=y_test-y_pred
```

```python
#1 Linearity check
plt.figure(figsize=(6,3))
sns.scatterplot(x=y_pred, y=y_test)
plt.title('Actual vs Predicted')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
#2 Homoscedasticity check
plt.figure(figsize=(6,3))
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red')
plt.title("Homoscedasticity: Residuals vs Predicted")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
plt.show()
```

```python
#3 Normality of residuals: Histogram and QQ Plot
import scipy.stats as stats
plt.figure(figsize=(6, 4))
sns.histplot(residuals, kde=True)
plt.title("Histogram of Residuals")
plt.show()

plt.figure(figsize=(6, 4))
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("QQ Plot of Residuals")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("Model Evaluation")
print('RMSE:', mean_squared_error(y_test,y_pred, squared=False))
print('RÂ² score:', r2_score(y_test, y_pred))
```

### As we can see the R square value is less than 0, which means model performance is too poor.

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=42)
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)

print("Decision Tree Regressor")
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("RÂ² score:", r2_score(y_test, y_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

print("Random Forest Regressor")
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("RÂ² score:", r2_score(y_test, y_pred))
```
